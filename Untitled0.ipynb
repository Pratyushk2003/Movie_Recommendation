```python
import numpy as np
import pandas as pd  # (our data in the form of csv and it is not easy to analyse data in csv therefore we use pandas dataframe wherein data is in the form of structured tables) 
import difflib # this is used because we need to take name of movies as input from user and if the user has done a spelling mistake then this difflib will find the closest match of a movie to that spelling given by user
from sklearn.feature_extraction.text import TfidfVectorizer # use to convert textual data into feature vecters (numerical values)
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('/content/Db_movies.csv')
```

```python
df.head()
```

```python
# Get the column names to remove
columns_to_remove = df.columns[23:1284]

# Remove the columns
df = df.drop(columns_to_remove, axis=1)
```

```python
df.columns
```

```python
df.shape
```

```python
selected_features = ['genres', 'keywords', 'tagline', 'cast'] # list
print(selected_features)
```

```python
# replacing the null values with null string
for feature in selected_features:
  df[feature] = df[feature].fillna('')
```

```python
# combining all the 5 selected features

combined_features=df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']    
```

```python
print(combined_features)
```

```python
# converting the text data to feature vectors
# cosine similarity algorithm works better on numerical data rather than textual data hence we convert textual data into numerical vlaues using Tfidf Vectors

vectorizer = TfidfVectorizer() #loading Tfidf Vectorizer into an instance called vectorizer
```

```python
feature_vectors = vectorizer.fit_transform(combined_features) #converts the textual data into numerical values
```

```python
print(feature_vectors)
```

```python
# getting the similarity score (or similarity confidence value) using cosine similarity

similarity= cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
# getting the movie name from the user

movie_name = input(' Enter your favourite movie name : ')
```

```python
# creating a list with all the movies given in the dataset
list_of_all_titles = df['title'].astype(str).tolist()
print(list_of_all_titles)
```

```python
# Replace empty strings with NaN
df['id'] = df['id'].replace('', pd.NA)

# Convert the column to integer
df['id'] = df['id'].astype(float).astype('Int64')
```

```python
df.head()
```

```python
# Convert the movie_name to a string if it's not already
movie_name = str(movie_name)

if isinstance(list_of_all_titles, list):
    close_matches = difflib.get_close_matches(movie_name, list_of_all_titles)
    print(close_matches)
else:
    print("list_of_all_titles should be a list")
```

```python
close_match=close_matches[0]  # first movie is the most similar movie name
print(close_match)
```

```python
df.reset_index(inplace=True)

# Print the DataFrame to verify the addition of the index column
df.head()
```

```python
# find the index of the movie with the title

index_of_the_movie = df[df.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python

# getting a list of similar movies

similarity_score = list(enumerate(similarity[index_of_the_movie])) #enumerate is basically used to carry out a loop {also find the index of the movie}
# the movies similar to iron man will have a high similarity score whereas movies different from iron man will have a low similarity score

print(similarity_score)
```

```python
# sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key= lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index

print('Movies Suggested for you : \n')

i=1

for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index = df[df.index == index]['title'].values[0]
  if(i<=30):
    print(i,'.',title_from_index)
    i+=1
```
